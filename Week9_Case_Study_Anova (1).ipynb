# ● Check whether all the zones generate the same amount of sales.

# Stating the hypothesis 

# Null Hypothesis (H0): Sales generated by all zones (A,B,C and D) are equal

# Alternate Hypothesis (H1): Sales generated by all zones (A,B,C and D) are not equal 

# Significance level is set as α= 0.05

```python
#Conducting One-way ANOVA Test for finding the correct hypothesis
from scipy import stats
Zone_A_Sales = sales_zone_data['Zone - A']
Zone_B_Sales = sales_zone_data['Zone - B']
Zone_C_Sales = sales_zone_data['Zone - C']
Zone_D_Sales = sales_zone_data['Zone - D']

α= 0.05
F_value_sales, p_value_sales = stats.f_oneway(Zone_A_Sales,Zone_B_Sales,Zone_C_Sales,Zone_D_Sales)
print('F test statistic value:',round(F_value_sales,3))
print('p-value for sales data:',round(p_value_sales,3))
```

```python
if p_value_sales < α :
    print("Reject null hypothesis which means that the Sales for all zones are not equal")
else:
    print("Accept null hypothesis which means that that the Sales for all zones are equal.")
```
